### <ê·¸ë¦¼ 2-8>ì˜ ë°ì´í„° (page 84)

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import numpy as np
import pandas as pd
import random

# ì„ì˜ ìˆ«ì ì‹œë“œ ì„¤ì •
random.seed(20) # ì„ì˜ë¡œ 20 ì§€ì • 
np.random.seed(21) # ì„ì˜ë¡œ 21 ì§€ì •

# ê³ ê°ID ë°°ì—´ 
# 100000ê³¼200000ì‚¬ì´ ëœë¤í•œ ì •ìˆ˜ 10ê°œë¥¼ ë½‘ì•„ ë°°ì—´ë¡œ ë°˜í™˜í•œ ê²ƒì„ user_idë¼ ì§€ì •
# ì´ë•Œ, 100000 í¬í•¨í•˜ë©° 200000 í¬í•¨í•˜ì§€ ì•ŠìŒ
user_id = np.random.randint(100000, 200000, 10)

# êµ¬ë§¤ ê°€ê²© ë¦¬ìŠ¤íŠ¸ 
# 100ì—ì„œ 1000 ì‚¬ì´ ê°’ì—ì„œ ë¬´ì‘ìœ„ë¡œ 10ê°œë¥¼ ë½‘ì•„ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜ (1000 ë¶ˆí¬í•¨)
price_list = random.sample(range(100, 1000), 10)
# ê° ìš”ì†Œì— 1000ì„ ê³±í•œ ë’¤ $ë¥¼ ë¶™ì—¬ ë¬¸ìì—´ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
price_list = ['$' + str(x*1000) for x in price_list]

# í• ì¸ ê¸ˆì•¡ ë¦¬ìŠ¤íŠ¸  
# 100ì—ì„œ 500 ì‚¬ì´ ê°’ì—ì„œ ë¬´ì‘ìœ„ë¡œ 10ê°œë¥¼ ë½‘ì•„ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜ (500 ë¶ˆí¬í•¨)
reduce_price_list = random.sample(range(100, 500), 10)

# ê° ìš”ì†Œì— 10ì„ ê³±í•œ ë’¤ $ë¥¼ ë¶™ì—¬ ë¬¸ìì—´ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
reduce_price_list = ['$' + str(x*10) for x in reduce_price_list]

# VIP ì—¬ë¶€ ë°°ì—´
# YëŠ” 0.3ì˜ ë½‘í í™•ë¥ ë¡œ, Nì€ 0.7ì˜ ë½‘í í™•ë¥ ë¡œ 10ê°œì˜ Yì™€ Nìœ¼ë¡œ êµ¬ì„±ëœ ë°°ì—´ ë°˜í™˜
vip_array = np.random.choice(["Y", "N"], size=10, p=[0.3, 0.7])

# ì„ì˜ ë°ì´í„° ë§Œë“¤ê¸° 
df = pd.DataFrame({
                    'ê³ ê°ID': user_id, 
                    'êµ¬ë§¤ë‚ ì§œ': pd.Timestamp('20210331'),
                    'êµ¬ë§¤ê°€ê²©': price_list,
                    'í• ì¸ê¸ˆì•¡': reduce_price_list,
                    'VIP': vip_array
                  })
df
```

```python
df.dtypes
```

```python
df["í• ì¸ê¸ˆì•¡"] = df["í• ì¸ê¸ˆì•¡"].str.replace("$", "").astype(int)
df
```

```python
# ë¦¬ìŠ¤íŠ¸ì— ìˆëŠ” ì—´ë§ˆë‹¤ 0.25ë§Œí¼ì˜ í–‰ì„ ìƒ˜í”Œ ì¸ë±ìŠ¤ë¥¼ ì¶”ì¶œí•´ í•´ë‹¹ ì¸ë±ìŠ¤ì— ìˆëŠ” ê°’ì„ np.nanìœ¼ë¡œ ë³€í™˜
for col in ['êµ¬ë§¤ë‚ ì§œ', 'êµ¬ë§¤ê°€ê²©', 'í• ì¸ê¸ˆì•¡', 'VIP']:
    df.loc[df.sample(frac=0.25).index, col] = np.nan
df
```
