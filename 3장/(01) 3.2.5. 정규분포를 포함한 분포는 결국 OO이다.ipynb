### 3.2.5.1 Î∂ÑÌè¨Îäî Í≤∞Íµ≠ Ìï®ÏàòÎã§. (Page 169)

```python
# ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞ 
import numpy as np
import math

# Í≤ΩÍ≥† ÌëúÏãú ÏßÄÏõÄ 
import warnings
warnings.filterwarnings('ignore')

# Í∑∏ÎûòÌîÑ Í¥ÄÎ†® ÎùºÏù¥Î∏åÎü¨Î¶¨  
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc

# matplotlib Ìè∞Ìä∏ ÏÑ§Ï†ï (Mac OSÏù∏ Í≤ΩÏö∞)
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# xÍ∞í Î≤îÏúÑ: -3ÏóêÏÑú 3ÍπåÏßÄ 0.05Ïùò Í∞ÑÍ≤©ÏúºÎ°ú Î∞∞Ïó¥ ÏÉùÏÑ± 
x = np.array(np.arange(-3,3, 0.05))
# yÍ∞í Î≤îÏúÑ: xÍ∞íÏùÑ Ìï¥Îãπ yÏãùÏóê ÎåÄÏûÖ
y = np.exp(-x**2)

# Í∑∏ÎûòÌîÑ ÎßåÎì§Í∏∞
plt.plot(x,y, color='green', linewidth=2)
plt.rcParams['figure.figsize'] = [11, 6]

# Ï†úÎ™© Î∞è XÏ∂ï, YÏ∂ï Ïù¥Î¶Ñ ÏÑ§Ï†ï 
plt.title("$y=e^{-x^2}$")
plt.xlabel('x')
plt.ylabel('y')

# Í∑∏ÎûòÌîÑ Î∞∞Í≤ΩÌôîÎ©¥ ÏÑ§Ï†ï 
plt.grid(alpha=.4, linestyle='--')

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
```

```python
# xÍ∞í Î≤îÏúÑ: -3ÏóêÏÑú 3ÍπåÏßÄ 0.05Ïùò Í∞ÑÍ≤©ÏúºÎ°ú Î∞∞Ïó¥ ÏÉùÏÑ± 
x = np.array(np.arange(-3,3, 0.05))

# ÏµúÏ¥àÏãù 
y = np.exp(-x**2)
# yÍ∞í Î≤îÏúÑ: xÍ∞íÏùÑ Ìï¥Îãπ ÌëúÏ§ÄÏ†ïÍ∑úÎ∂ÑÌè¨ÏãùÏóê ÎåÄÏûÖ
y_std_normal = 1/(math.sqrt(2*math.pi))*np.exp(-(1/2.)*x**2)

# Í∑∏ÎûòÌîÑ ÎßåÎì§Í∏∞
plt.plot(x,y, label='ÏµúÏ¥àÏãù', color='green', linewidth=2)
plt.plot(x,y_std_normal, label="ÌëúÏ§ÄÏ†ïÍ∑úÎ∂ÑÌè¨", linestyle='--', color="blue", linewidth=2)
plt.rcParams['figure.figsize'] = [11, 6]

# Ï†úÎ™© Î∞è XÏ∂ï, YÏ∂ï Ïù¥Î¶Ñ ÏÑ§Ï†ï 
plt.title("ÌëúÏ§ÄÏ†ïÍ∑úÎ∂ÑÌè¨, $\mu=0, \sigma=1$")
plt.xlabel('x')
plt.ylabel('y')

# legend
plt.legend()

# Í∑∏ÎûòÌîÑ Î∞∞Í≤ΩÌôîÎ©¥ ÏÑ§Ï†ï 
plt.grid(alpha=.4, linestyle='--')

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
```

```python
## xÍ∞í Î≤îÏúÑ: -3ÏóêÏÑú 3ÍπåÏßÄ 0.05Ïùò Í∞ÑÍ≤©ÏúºÎ°ú Î∞∞Ïó¥ ÏÉùÏÑ± 
x = np.array(np.arange(-6,6, 0.05))
# mu=0, sigma = 1 (ÌëúÏ§ÄÏ†ïÍ∑úÎ∂ÑÌè¨)
y = 1/(math.sqrt(2*math.pi))*np.exp(-(1/2.)*x**2)
# mu=2, sigma = 1
y_mu = 1/(math.sqrt(2*math.pi))*np.exp(-(1/2.)*(x-2)**2)
# mu=0, sigma = 2
y_sigma = 1/(2*math.sqrt(2*math.pi))*np.exp(-(1/2.)*(x/2.)**2)

# Îëê Í∞ú Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ 
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x,y, label="$\mu$=0, $\sigma$=1", color="blue", linewidth=2, linestyle='--')
ax1.plot(x,y_mu, label="$\mu$=2, $\sigma$=1", color="gray", linewidth=2)
ax1.title.set_text("$\mu$Í∞íÏóê Îî∞Î•∏ Ï†ïÍ∑úÎ∂ÑÌè¨")
ax1.grid(alpha=.4, linestyle='--') # Í∑∏ÎûòÌîÑ Î∞∞Í≤ΩÌôîÎ©¥ ÏÑ§Ï†ï
ax1.legend() # Î≤îÎ°Ä ÎÑ£Í∏∞ 

ax2.plot(x,y, label="$\mu$=0, $\sigma$=1", color="blue", linewidth=2, linestyle='--')
ax2.plot(x,y_sigma, label="$\mu$=0, $\sigma$=2", color="gray", linewidth=2)
ax2.title.set_text("$\sigma$Í∞íÏóê Îî∞Î•∏ Ï†ïÍ∑úÎ∂ÑÌè¨")

ax2.grid(alpha=.4, linestyle='--') # Í∑∏ÎûòÌîÑ Î∞∞Í≤ΩÌôîÎ©¥ ÏÑ§Ï†ï
ax2.legend() # Î≤îÎ°Ä ÎÑ£Í∏∞

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
plt.rcParams['figure.figsize'] = [11, 6]
```

### 3.2.5.2. Î∂ÑÌè¨Îäî Î™®Îç∏(Î™®Ìòï)Ïù¥Îã§
- page 175

```python
import scipy.stats
from scipy import stats
plt.style.use('ggplot') # plot Ïä§ÌÉÄÏùº ÏÑ§Ï†ï 

x = np.linspace(-3.2, 3.2, 1000) # [-3.2 3.2]Î≤îÏúÑÏóêÏÑú Í∞úÏàòÍ∞Ä 1000Í∞úÍ∞Ä ÎêòÎèÑÎ°ù Í∞í ÏÑ§Ï†ï
iq = stats.norm(0, 1) # ÌèâÍ∑†:0, ÌëúÏ§ÄÌé∏Ï∞®:1Ïù∏ pdf

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ 
plt.plot(x, iq.pdf(x), 'gray', lw=2)
plt.rcParams['figure.figsize'] = [11, 6]

# Í∑∏ÎûòÌîÑÏóê mu, sigma Í∏ÄÏûê Ï∂îÍ∞Ä 
plt.text(-3.2, -0.02, "$\mu - 3\sigma$", size=20)
plt.text(-2.2, -0.02, "$\mu - 2\sigma$", size=20)
plt.text(-1.2, -0.02, "$\mu - \sigma$", size=20)
plt.text(0, -0.02, "$\mu$", size=20)
plt.text(0.8, -0.02, "$\mu + \sigma$", size=20)
plt.text(1.8, -0.02, "$\mu + 2\sigma$", size=20)
plt.text(2.8, -0.02, "$\mu + 3\sigma$", size=20)

# ÌëúÏ§ÄÌé∏Ï∞®Ïóê Îî∞Î•∏ ÌôïÎ•†Í∞í Í≥ÑÏÇ∞: ÏòàÏãú) Prob(X<1) - Prob(X<-1) 
std_1 = round(scipy.stats.norm(0, 1).cdf(1) -
              scipy.stats.norm(0, 1).cdf(-1), 3) # Prb(X<1) - Prob(X<-1)
std_2 = round(scipy.stats.norm(0, 1).cdf(2) -
              scipy.stats.norm(0, 1).cdf(-2), 3) # Prb(X<2) - Prob(X<-2)
std_3 = round(scipy.stats.norm(0, 1).cdf(3) -
              scipy.stats.norm(0, 1).cdf(-3), 3) # Prb(X<3) - Prob(X<-3)

# Í∑∏ÎûòÌîÑÏóê Í≥ÑÏÇ∞Îêú ÌôïÎ•† ÌÖçÏä§Ìä∏ Ï∂îÍ∞Ä 
plt.text(-0.2, 0.25, std_1, size=20)
plt.text(-0.2, 0.06, std_2, size=20)
plt.text(-0.2, 0.01, std_3, size=20)

# ÌôîÏÇ¥Ìëú Ï∂îÍ∞Ä 
# scipy.stats.norm(0, 1).pdf(-1): xÍ∞Ä -1ÏùºÎïå ÌëúÏ§ÄÏ†ïÍ∑úÎ∂ÑÌè¨Î•º Í∞ÄÏßÑ yÍ∞í
plt.annotate("", 
             xy=(-1, scipy.stats.norm(0, 1).pdf(-1)), 
             xytext=(1, scipy.stats.norm(0, 1).pdf(1)), 
             arrowprops=dict(arrowstyle="<->",
                             connectionstyle="arc3", 
                             color='green', 
                             lw=3))
plt.annotate("", 
             xy=(-2, scipy.stats.norm(0, 1).pdf(-2)), 
             xytext=(2, scipy.stats.norm(0, 1).pdf(2)), 
             arrowprops=dict(arrowstyle="<->",
                             connectionstyle="arc3", 
                             color='blue', 
                             lw=3))
plt.annotate("", 
             xy=(-3, scipy.stats.norm(0, 1).pdf(-3)), 
             xytext=(3, scipy.stats.norm(0, 1).pdf(3)), 
             arrowprops=dict(arrowstyle="<->",
                             connectionstyle="arc3", 
                             color='red', 
                             lw=3))

# Í∑∏ÎûòÌîÑ ÏÉâ ÏßÄÏ†ï 
colors = ['red', 'blue', 'green']
colors = colors + list(reversed(colors))

# Ï°∞Í±¥Ïóê Îî∞Îùº Îã§Î•∏ ÏÉâÏúºÎ°ú ÏßÄÏ†ï 
mean, std = 0, 1
for i, color in zip(range(-3, 3), colors):
    # -3 Îπ®Í∞ÑÏÉâ, -2 ÌååÎûÄÏÉâ, -1 Ï¥àÎ°ùÏÉâ, 0 Ï¥àÎ°ùÏÉâ, 1 ÌååÎûÄÏÉâ, 2 Îπ®Í∞ÑÏÉâ 
    low = mean + i * std
    high = mean + (i + 1) * std
    px = x[np.logical_and(x >= low, x <= high)]
    plt.fill_between(
        px,
        iq.pdf(px),
        color=color, 
        alpha=0.4, # Ìà¨Î™ÖÎèÑ 
        linewidth=0 # ÏÑ† ÎëêÍªò
    )
```
